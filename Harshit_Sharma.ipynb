```python
import pandas as pd

import numpy as np

raw_data = pd.read_csv('Downloads\loan.csv') 
print(raw_data.head()) # giving quick a look on data



useful_data  = raw_data.loc[: , ['avg_cur_bal','dti','emp_length','grade','home_ownership','int_rate','pub_rec_bankruptcies','purpose','term']]# keeping only useful columns
useful_data.head() #checking data

print(useful_data.isnull().sum())

useful_data = useful_data[~(useful_data['pub_rec_bankruptcies']).isnull()]  #droping null values

useful_data.isnull().sum().value_counts()

#useful_data.info()  # sanity check
print(useful_data.loc[:,'purpose'].value_counts())
print(useful_data.loc[:,'pub_rec_bankruptcies'].value_counts())


# (avg_cur_bal = 39020) one thing is clear from data 39020 folks have no avg_cur_bal likely to default

# Major chunk is debt_consolidation who liklely to default near future when amount becomes mamonth (reducing amount)

# pub_rec_bankruptcies is factor determine their past and likely default in future (risky applicants)

print(useful_data.loc[:,['emp_length','grade']].value_counts().sort_values())
useful_data.loc[:,['dti','home_ownership']].value_counts().sort_values(ascending = False)

#most likely emp with 10+ year experiance are mostly bad graded (risky applicants)

#folks on rent has a max dti ratio (reducing amount)

```
